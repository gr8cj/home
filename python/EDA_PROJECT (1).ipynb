```python
def uni_numerical_analysis(dataframe, column):
    sns.set(style="darkgrid")
    plt.figure(figsize=(18,10))

    plt.subplot(1,2,1)
    sns.boxplot(data=dataframe, x=column, orient='v').set(title='boxplot')
    plt.subplot(1,2,2)
    sns.histplot(dataframe[column], kde=True).set(title="histplot")

    plt.show()
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
for i in numerical_columns:
    uni_numerical_analysis(df_app, i)
```

```python
def uni_categorical_analysis(dataframe, column):
    plt.figure(figsize=(16, 10))
    sns.barplot(x=dataframe[column].value_counts(), y=dataframe[column].value_counts().index)
    plt.title(f" frequency of {column}", fontsize=15)
    plt.show()
```

```python
for i in categorical_columns:
    uni_categorical_analysis(df_app, i)
```
